```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("WordCount").master("yarn").getOrCreate()
```

```python
spark
```

```python
data = [
"Goku Vegeta Gohan",
"Goku Frieza Goku",
"Vegeta Goku Gohan Frieza",
"Gohan Frieza Goku Goku"
]
```

```python
# Create a text file from the data
with open("names.txt", "w") as f:
  for line in data:
    f.write(line + "\n")
```

- In terminal, after we create file on local disk. we use following command to put it on hdfs 
    - hadoop fs -put /names.txt /tmp/
- then we can proceed with following steps

```python
hdfs_path = "/tmp/names.txt"
```

```python
rdd1 = spark.sparkContext.textFile(hdfs_path)

rdd1.collect()
```

```python
# transformation
words = rdd1.flatMap(lambda line: line.split(" "))
word_counts = words.map(lambda word: (word, 1)).reduceByKey(lambda a, b: a + b)

# action
result = word_counts.collect()

# result
print(result)
```

```python
spark.stop()
```
